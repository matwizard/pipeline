# Load the workbooks into (collections of) pandas dataframes
'read_excel' creates an 'ordered dict' collection of dataframes, one for each worksheet in the workbook

```python
od100 = pd.read_excel("Data_CGR100.xlsx", sheet_name=None)
od500 = pd.read_excel("Data_CGR500.xlsx", sheet_name=None)
od1000 = pd.read_excel("Data_CGR1000.xlsx", sheet_name=None)
```

```python
od1000.keys()
```

# Convert and save to python pickle fines

```python
pk.dump(od100,open("od100.p", 'wb'))
pk.dump(od500,open("od500.p", 'wb'))
pk.dump(od1000,open("od1000.p", 'wb'))
```

```python
nod100 = pk.load(open("od100.p", 'rb'))
nod500 = pk.load(open("od500.p", 'rb'))
nod1000 = pk.load(open("od1000.p", 'rb'))
```

```python
arr100 = np.stack([np.array(i) for i in od100.values()][:-1])
arr500 = np.stack([np.array(i) for i in od500.values()][:-1])
arr1000 = np.stack([np.array(i) for i in od1000.values()][:-1])
```
